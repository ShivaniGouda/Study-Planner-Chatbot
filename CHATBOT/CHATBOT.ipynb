```python
import requests

# HF_API_KEY = "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"
API_URL = "https://api-inference.huggingface.co/models/HuggingFaceH4/zephyr-7b-beta"

headers = {
    "Authorization": f"Bearer {HF_API_KEY}",
    "Content-Type": "application/json"
}

def get_study_plan(topic):
    prompt = f"<|user|>\nI want to study {topic}. Please create a 4-week study plan with weekly goals and resource recommendations.\n<|assistant|>\n"
    
    payload = {
        "inputs": prompt,
        "parameters": {
            "temperature": 0.7,
            "max_new_tokens": 700,
            "return_full_text": False
        }
    }
    
    response = requests.post(API_URL, headers=headers, json=payload)
    if response.status_code != 200:
        raise Exception(f"Failed: {response.status_code}\n{response.text}")
    return response.json()[0]['generated_text']

# Example usage
user_input = input("What would you like to study? ")
plan = get_study_plan(user_input)
print("\n--- Your Study Plan ---\n")
print(plan)
```
